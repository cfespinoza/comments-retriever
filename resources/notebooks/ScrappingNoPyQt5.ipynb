## Extract content from ElMundo's article

```python
from lxml import html as htmlRenderer
urlNoticiaElMundo = "https://www.elmundo.es/espana/2019/09/07/5d739634fc6c83713b8b46bd.html"
responseNoticia = requests.get(urlNoticiaElMundo)
renderedPage = htmlRenderer.fromstring(responseNoticia.text)
# data-section="articleBody"
bodyParts = renderedPage.xpath("//div[@data-section='articleBody']//p")
bodyParts
```

```python
contentArr = []
for p in bodyParts:
    contentArr.append(p.text_content())
contentArr
```

## Extract sections from home page

```python
# 2019/01/01
from lxml import html as htmlRenderer
urlHemeroteca = "https://www.elmundo.es/elmundo/hemeroteca/2019/01/01/m/index.html"
responseNoticia = requests.get(urlHemeroteca)
renderedPage = htmlRenderer.fromstring(responseNoticia.text)
references = renderedPage.xpath("//ul[@class='nav-tabs js-mainNavigation']//li//a/@href")
filteredSections = list(dict.fromkeys([section for section in references if not section.startswith("http")]))
finalSections = list(dict.fromkeys([section.split(".")[0] for section in filteredSections]))
len(finalSections)
```
