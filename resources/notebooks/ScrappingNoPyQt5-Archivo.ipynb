```python
from bs4 import BeautifulSoup
import requests
```

```python
urlPortada="https://www.20minutos.es/archivo/2019/01/7/"
responsePortada = requests.get(urlPortada)
```

```python
responsePortada.text
```

```python
from lxml import html as htmlRenderer
renderedPage = htmlRenderer.fromstring(responsePortada.text)
auxLinks = renderedPage.xpath("//a/@href")
```

```python
auxLinks
```

## Getting Urls From Home Page

```python
soup = BeautifulSoup(responsePortada.text)
for link in soup.find_all('a'):
    print(link.get("href"))
```
