```python
import csv
import sys
import requests
import json
from lxml import html as htmlRenderer
csv.field_size_limit(sys.maxsize)
```

```python
csvFileIn = "/home/cflores/Descargas/20minutos-01012019-15092019-contents.csv"
csvFileOut = "/home/cflores/Descargas/20minutos-01012019-15092019-contents-categorias.csv"

data_layer_key = "dataLayer"
categoria_key = "categoria"
subcategoria_key = "subcategoria"
```

```python
def get_data_layer(url=None, rendered_html=None):
    scriptText = ""
    for element in rendered_html.iter('script'):
        scriptText = element.text_content()
        if data_layer_key in scriptText:
            print(" data_layer_key has been found")
            break
    if not scriptText:
        return {}
    
    data_layer_json_str = scriptText.split('dataLayer = [')[-1].rsplit('];', 1)[0].strip()
    try:
        data_layer = json.loads(data_layer_json_str)
        return data_layer
    except Exception as e: 
        print(" parsing json has failed.")
        print("-------------------------------")
        print(data_layer_json_str)
        print("-------------------------------")
        print(str(e))
        return {}
    
```

```python
with open(csvFileIn) as csvIn:
    with open(csvFileOut, 'w') as csvOut:
        reader = csv.reader(csvIn)
        writer = csv.writer(csvOut)
        rowNumber = 0
        for row in reader:
            if rowNumber == 0:
                headers = row + [categoria_key, subcategoria_key]
                writer.writerow(headers)                
            else:
                url = row[0]
                html = requests.get(url)
                rendered_html = htmlRenderer.fromstring(html.text)
                data_layer = get_data_layer(url, rendered_html)
                category = data_layer.get(categoria_key, "")
                subcategory = data_layer.get(subcategoria_key, "")
                writer.writerow(row + [category, subcategory])
            rowNumber = rowNumber + 1
```
